```python
import os
from dotenv import load_dotenv

from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()

#LLM
llm = ChatOpenAI(model = "gpt-4o-mini", temperature = 0)

# Output Parser (converts AIMessage -> string)
parser = StrOutputParser()

# Prompts
explain_prompt = ChatPromptTemplate.from_template(
    "Expalin {topic} in detail."
) 

summary_prompt = ChatPromptTemplate.from_template(
    "Summarise the following text in three lines:\n\n{input}"
)

bullet_prompt = ChatPromptTemplate.from_template(
    "Convert the following summary into bullet points:\n\n{input}"
)

# Proper 3-step chain
chain = (
    explain_prompt
    | llm
    | parser
    | summary_prompt
    | llm
    | parser
    | bullet_prompt
    | llm
    | parser
)

# run
response = chain.invoke({"topic": "Machine Learning"})

print(response)
```

```python
# full langchin program

import os 
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# load .env
load_dotenv()

# llm
llm = ChatOpenAI(
    model = "gpt-4o-mini",
    temperature = 0
)

# output_parser
parser = StrOutputParser()

# prompts
explain_prompt = ChatPromptTemplate.from_template(
    "Expalin {topic} in detail."
) 

summary_prompt = ChatPromptTemplate.from_template(
    "Summarise the following text in 3 lines:\n\n{input}"
)

bullet_prompt = ChatPromptTemplate.from_template(
    "Convert the following text in bullet points:\n\n{input}"
)

# Chain Creation
chain = (explain_prompt 
        | llm
        | parser
        | summary_prompt
        | llm
        | parser
        | bullet_prompt
        | llm
        | parser
        )

# Run 
response  = chain.invoke({"topic": "Claude AI"})

# print
print(response)
```
