## Import required libraries  
- LangChain
- Ollama Embeddings
- FIASS Vector Database
- ChromaDB vector Database


```python
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

## 3. Create Embedding Model  
#### Embeddings convert text into numerical vectors that capture semantic meaning

```python
embeddings = OllamaEmbeddings(
    model = "nomic-embed-text"
)
```

## 4. Load a document

```python
loader = PyMuPDFLoader("../Gen_ai_data/sample_ai_document.pdf")
docs = loader.load()
len(docs)
```

## 5. Split The document into chunks

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)

chunks = splitter.split_documents(docs)

len(chunks)
```

# Par1: FIASS Vector Database

## FIASS is a fast similiarity search library developed by data 



## 6. Create FAISS Vector Storage

```python
faiss_db = FAISS.from_documents(
    chunks, 
    embeddings
)
```

## 7. Search Similar Documents with Fiass

```python
query = "What is Artificial Intelligence?"

results = faiss_db.similarity_search(query)

results[0].page_content
```

## 8. Convert FAISS into Retriever

```python
retriever = faiss_db.as_retriever()

retriever.invoke("Explain Artificial Intelligence")
```

# Part 2: ChromaDB Vector Database
## ChromaDB is a lightweight vector DataBase designed for AI application

## 9. Create Chroma Vector Store

```python
chroma_db = Chroma.from_documents(
    documents = chunks,
    embedding = embeddings,
    persist_directory = '../Genai_project_data/chroma_db'
)
```

## 10. Search Documents with ChromaDB

```python
query = "Explain Artificial Intelligence"

results = chroma_db.similarity_search(query)

results[0].page_content
```

## 11. Create Retriever from ChromaDB 

```python
retriever = chroma_db.as_retriever()

retriever.invoke("What is Machine Learning")
```
