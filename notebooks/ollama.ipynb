# Ollama Genai Without memory

```python
import ollama 
print("ollama simple chat (No Memory)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() in ['exit', 'quit']:
        break

    response = ollama.generate(
        model = 'llama3.2',
        prompt = user_input
    )

    print("Bot: ", response['response'])
```

# Ollama Genai With memory

```python
import ollama 

# Store conversation history

messages = [{'role':'system', 'content':'You are an helpful assistant.'}]

print('Ollama LLaMA 3.2 Chatbot  Started.')
print("Type 'exit' or 'quit' to stop.\n")

try:
    while True:
        user_input = input("You: ")

        if user_input.lower() in ['exit', 'quit']:
            print("chat ended.")
            break
        # add user message
        messages.append(
            {
                'role':'user',
                'content':user_input
            }
        ) 

        # call Ollama Chat API
        response = ollama.chat(
            model = "llama3.2",
            messages = messages
        )

        bot_reply = response['message']['content']

        # add assistant reply to history
        messages.append({
            "role":"assistant",
            'content':bot_reply
        })

        print("Bot: ", bot_reply)

except KeyboardInterrupt:
    print("\n Chat Interrupted manually")
```
