## 1. from langchain_ollama import OllamaEmbedding, ChatOllama


```python
from langchain_ollama import OllamaEmbeddings, ChatOllama
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
```

## 2. CREATE EMEDDING MODEL 


```python
embedding = OllamaEmbeddings(
    model="nomic-embed-text"
)
```

## 3. embed a single text 


```python
text ="Artificial Inteligence is tranforming a world"
vector = embedding.embed_query(text)
print("Vector length:" ,len(vector))
print(vector[:10])
```

## 4. embed in multiple documents 


```python
documents=[
    "Artificial Inteligence is tranforming industires",
    "Machine Learning enables systems to learn from data",
    "Deep learning uses neural networks"
]
vector = embedding.embed_documents(documents)
print(len(vector))
```

## 5. load pdf documents


```python
loader = PyMuPDFLoader("../Gen_ai_data/sample_ai_document.pdf")
docs = loader.load()
print(len(docs))
```

## 8. split documents into chunks


```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)

chunks = splitter.split_documents(docs)

print(len(chunks))
```

## 9. Create embeddings for chunks

```python
vectors = embedding.embed_documents(
    [chunk.page_content for chunk in chunks]
)

print(len(vectors))
```

## 10. Store Embeddings in FIASS Vector Database

```python
vectorstore = FAISS.from_documents(
    chunks, 
    embedding
)
```

## 11. Search Similar Documents 

```python
query = "What is Artificial Intelligence?"

results = vectorstore.similarity_search(query)

print(results[0].page_content)
```

## 12. Ask Question Using an LLm 

```python
llm = ChatOllama(
    model = "llama3.2"
)

response = llm.invoke("Explain Artificial Intelligence in simple Words")

print(response.content)
```
