### Basic Character Text Splitter Example

```python
text = """Artificial Intelligence is transforming industrires. 
Machine Learning enables systems to learn from data.
deep learning uses neural networks to detect Patterns."""

splitter = CharacterTextSplitter(
    chunk_size = 50,
    chunk_overlap = 10
)

chunks = splitter.split_text(text)

chunks
```

### Recursive Character Text Splitter (Most Common in Rag)

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20
)

chunks = splitter.split_text(text)
chunks
```

### Token Text Splitter (Token Aware)

```python
splitter = TokenTextSplitter(
    chunk_size = 30,
    chunk_overlap = 5
)

chunks = splitter.split_text(text)

chunks
```

### Load and split a PDF Document

```python
loader = PyMuPDFLoader("../Gen_ai_data/sample_ai_document.pdf")

docs = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)

pdf_chunks = splitter.split_documents(docs)

len(pdf_chunks)
```

```python
loader = CSVLoader("../Gen_ai_data/sample_sales_data.csv")

docs = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap = 20
)

csv_chunks = splitter.split_documents(docs)

len(csv_chunks)
```

### This is Working

```python
from pptx import Presentation
from langchain_core.documents import Document

prs = Presentation("C:/Users/bathr/Desktop/Cybrom_Class_Data/Genai_projects/Gen_ai_data/sample_ai_presentation.pptx")

docs = []

for slide in prs.slides:
    text = ""
    for shape in slide.shapes:
        if hasattr(shape, "text"):
            text += shape.text + "\n"

    docs.append(Document(page_content=text))

print(len(docs))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=30
)

ppt_chunks = splitter.split_documents(docs)

print(len(ppt_chunks))
```

### Load and split a web documents

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"

loader = WebBaseLoader(url)

docs = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size = 800,
    chunk_overlap = 100
)

web_chunks = splitter.split_documents(docs)

len(web_chunks)
```
