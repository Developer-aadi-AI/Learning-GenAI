# OpenAI Generative ai with memory

```python
from dotenv import load_dotenv
from openai import OpenAI
import os
# load env variables
load_dotenv()

# Create OpenAI Client
client = OpenAI(api_key = os.getenv('openai_api_key'))

# Store conversation History
messages = [{'role': 'system', 'content': 'you are a helpful assistant.'}]

print('Chatbot started. Type "exit" or "quit" to stop. \n' )

try:
    while True:
        user_input = input('You: ')

        if user_input.lower() in ['exit', 'quit']:
            print('Chat ended.')
            break


        # add users message
        message.append({'role':'user','content':user_input})

        # Get response from OpenAI
        response = client.chat.completions.create(
            model = 'gpt-4o-mini',
            messages = message,
            temperature = 0.3
        )

        bot_reply = response.choices[0].message.content

        # add assistant reply to memory
        messages.append({'role':'assistant', 'content': bot_reply})

        print('Bot: ', bot_reply)

except KeyboardInterrupt:
    print("\nChat interrupted Manually.")

    
```

# OpenAI Generative AI Without memory

```python
from dotenv import load_dotenv
from openai import OpenAI
import os
# load env variables
load_dotenv()

# Create OpenAI Client
client = OpenAI(api_key = os.getenv("openai_api_key"))

print('Chatbot started. (No memory)\n' )

while True:
    user_input = input("You: ")

    if user_input.lower() in ['exit', 'quit']:
        break

    response = client.responses.create(
        model = "gpt-4o-mini",
        input = user_input
    )

    print("Botendra:", response.output[0].content[0].text)
```
