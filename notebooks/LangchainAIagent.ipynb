# Calculator tool

```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_ollama import ChatOllama
```

# 1.Define tool

```python
@tool
def calculator(expr: str) -> str:
    """Evaluate a mathematical expression."""
    return str(eval(expr))
```

# 2. Load LLM

```python
llm = ChatOllama(model = "llama3.2")
```

# 3. Create agent

```python
agent = create_agent(
    model = llm,
    tools = [calculator],
    system_prompt = "You are a helpful assistant"
)
```

# 4. Invoke

```python
response = agent.invoke({"messages": [{"role":"user", "content":"24*4+10"}]})
print(response)
```

# 5. Print Answer Only

```python
print(response["messages"][-1].content)
```

# Word Count Tool

```python
from langchain.tools import tool
from langchain.agents import create_agent
from langchain_ollama import ChatOllama
```

```python
# 1. Define Tool
@tool
def word_count(text:str)->str:
    """Count the number of words in a given text."""
    return str(len(text.split()))
```

# 2. Load LLm

```python
llm = ChatOllama(model="llama3.2")
```

# 3. Create agent

```python
agent = create_agent(
    model = llm,
    tools = [word_count],
    system_prompt = "Use the word_count tool when user asks to count words. Return only the number."
)
```

# 4. Invoke

```python
response = agent.invoke({
    "messages":[{"role":"user", "content":"Count words in: Langchain is very powerful"}]
})
```

# 5. Print Only Answer

```python
print(response["messages"][-1].content)
```
